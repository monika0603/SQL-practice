# Converting string to numeric values

```python
df = pd.DataFrame({'col_one':['1.1','2.2','3.3'],
                   'col_two':['4.4','5.5','6.6'],
                   'col_three':['7.7','8.8','-']})
df
```

```python
df.astype({'col_one':float, 'col_two':float}).dtypes
```

```python
# By default to_numeric function fills non-numeric numbers with NaN
df = pd.to_numeric(df.col_three, errors='coerce')
df
```

```python
# You can use fillna(0) method to replace NaN with 0
```

# Change all string values to numeric types for a df

```python
df = pd.DataFrame({'col_one':['1.1','2.2','3.3'],
                   'col_two':['4.4','5.5','6.6'],
                   'col_three':['7.7','8.8','-']})
df
```

```python
df = df.apply(pd.to_numeric, errors='coerce').fillna(0)
df
```

```python
df.dtypes
```

# Aggregate by multiple functions

```python
df = pd.read_csv('./titanic.csv')
```

```python
df
```

```python
df.groupby('Embarked')['Age'].agg(['sum','count']).reset_index().head()
```

# Combine the output of an aggregation with a DF

```python
# group by reduces the length of the input data, however by using transform method we 
# can group by but still have the same length
df['count'] = df.groupby('Embarked')['Age'].transform('count')
df
```

# Style a DataFrame 

```python
format_dict = {'Date':'{:%m/%d/%y}', 'Close':'${:.2f}', 'Volume':'{:,}'}
```

```python
# This will display columns defined for formating
df.style.format(format_dict)
```

```python
df.style.format(format_dict)\
  .hide_index()\
  .highlight_min('Close', color='red')\
  .highlight_max('Close', color='lightgreen')
```

```python
df.style.format(format_dict)\
  .hide_index()\
  .background_gradient(subset='Volume', cmap='Blues')
```

```python
df.style.format(format_dict)\
  .hide_index()\
  .bar('Volume', color='lightblue', align='zero')\
  .set_option('Stock prices from October 2016')
```
